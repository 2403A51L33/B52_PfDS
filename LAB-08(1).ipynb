```python
import pandas as pd
import matplotlib.pyplot as plt
import os

filename = 'student_marks.csv'

if not os.path.exists(filename):
    print(f"Creating dummy data file: {filename}")
    data = {'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
            'Average Marks': [85, 92, 78, 88, 95]}
    dummy_df = pd.DataFrame(data)
    dummy_df.to_csv(filename, index=False)

df = pd.read_csv(filename)

if 'Subject' not in df.columns or 'Average Marks' not in df.columns:
    raise ValueError("Dataset must contain 'Subject' and 'Average Marks' columns.")

plt.figure(figsize=(10, 6))
bars = plt.bar(df['Subject'], df['Average Marks'], color='skyblue', edgecolor='black')

plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.title('Average Marks of Students by Subject', fontsize=14)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, f'{yval:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

max_subject = df.loc[df['Average Marks'].idxmax()]
min_subject = df.loc[df['Average Marks'].idxmin()]

print(f"Easiest Subject: {max_subject['Subject']} with average marks of {max_subject['Average Marks']}")
print(f"Toughest Subject: {min_subject['Subject']} with average marks of {min_subject['Average Marks']}")
```
