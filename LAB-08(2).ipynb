```python
import pandas as pd
import matplotlib.pyplot as plt
import random

df = pd.read_csv('sales_category.csv')

if 'Category' not in df.columns or 'Sales' not in df.columns:
    raise ValueError("Dataset must contain 'Category' and 'Sales' columns.")

colors = ['#%06X' % random.randint(0, 0xFFFFFF) for _ in range(len(df))]

plt.figure(figsize=(10, 6))
bars = plt.bar(df['Category'], df['Sales'], color=colors, edgecolor='black')

plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.title('Sales Distribution by Product Category', fontsize=14)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, f'{yval:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

best_seller = df.loc[df['Sales'].idxmax()]
print(f"Best-Selling Category: {best_seller['Category']} with sales of {best_seller['Sales']}")
```
