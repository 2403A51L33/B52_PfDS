```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': pd.date_range(start='2022-01-01', periods=12, freq='ME'),
    'Close': [150, 155, 160, 158, 162, 165, 170, 168, 172, 175, 180, 185]
}
df = pd.DataFrame(data)

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='navy')
plt.title('Apple Stock Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
